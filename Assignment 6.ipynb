# Question 1: Define Object Oriented Programming Language?
# Ans :
Object-oriented programming (OOP) is define as a type of computer programming (software design) in which programmers define the data type of a data structure, and also the types of operations (functions) that can be applied to the data structure.
In this way, the data structure becomes an object that includes both data and functions. In addition, programmers can create relationships between one object and another. For example, objects can inherit characteristics from other objects.

# Question 3: Differentiate between function and method?
# Ans :
Method and a function are the same, with different terms. A method is a procedure or function in object-oriented programming.
A function is a piece of code that is called by name. It can be passed data to operate on (i.e. the parameters) and can optionally return data (the return value). All data that is passed to a function is explicitly passed. A method is a piece of code that is called by a name that is associated with an object.

# Question 4: Define the following terms:
1. Class
2. Object
3. Attribute
4. Behavior
# Ans :

# 1. Class :

A category of objects. The class defines all the common properties of the different objects that belong to it.
# 2. Object :

A self-contained entity that consists of both data and procedures to manipulate the data.
# 3. Attribute :

Attributes are the variables that are defined inside a class.
# 4. Behavior :

Behaviors are the methods corresponding to the behavior of real world objects.

```python
# Question 5: Write a code in python in which create a class named it Car which have 5 attributes such like (model, color and name etc.) and 3 methods. And create 5 object instance from that class.

class Car():

    def __init__(self, name, model, color, company, province):
        self.name = name
        self.model = model
        self.color = color
        self.company = company
        self.province = province
        

    def available(self):
        print("\nAvailable Car :-\n")
        print(f"Name     : {self.name}")
        print(f"Model    : {self.model}")
        print(f"Color    : {self.color}")
        print(f"Company  : {self.company}")
        print(f"Province : {self.province} \n")
        

    def updatename(self, newname):
        self.name = newname
    def updatecolor(self, newcolor):
        self.color = newcolor
    def updatemodel(self,newmodel):
        self.model = newmodel


c1 = Car("Acura", 2018, "White", "Honda Motor", "Sindh")
c2 = Car("Alfa Romeo", 20019, "Blue", "Fiat Chrysler", "Punjab")
c3 = Car("Audi", 2020, "Black", "Volkswagen", "Balochistan")
c4 = Car("BMW", 2017, "Red", "BMW Group", "Sindh")
c5 = Car("Chevrolet", 2018, "White", "General Motors", "Punjab")


c1.available()
print("\n---------------------------------------------------------------------------------------------------\n")
c2.available()
print("\n---------------------------------------------------------------------------------------------------\n")
c3.available()
print("\n---------------------------------------------------------------------------------------------------\n")
c4.available()
print("\n---------------------------------------------------------------------------------------------------\n")
c5.available()

print("\n******************************* For Updating These Car Properties ***********************************\n")
c1.available()
print("\n******************************* After Updating Above Car Properties *********************************\n")
c1.updatename("Land Cruiser")
c1.updatecolor("Black")
c1.updatemodel("2015")
c1.available()
```
